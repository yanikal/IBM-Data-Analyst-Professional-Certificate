```python
#Normalize the column “Weight” so that the values range from 0 to 1 
#df['weight'].plot(kind = 'bar')

# copy the data
df_max_scaled = df.copy()
  
# apply normalization techniques on Column 1
column = 'weight'
df_max_scaled[column] =(df_max_scaled[column]-df_max_scaled[column].min())/(df_max_scaled[column].max()-df_max_scaled[column].min())
  
# view normalized data
display(df_max_scaled)
```

```python
# We will use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins

pd.cut(df['horsepower'], bins=3).value_counts()
```

```python
#We would like 3 bins of equal size bandwidth so we use numpy's linspace(start_value, end_value, numbers_generated function.
import numpy as np
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
## we set the group names
group_names = ['Low', 'Medium', 'High']
```

```python
#We apply the function "cut" the determine what each value of "df['horsepower']" belongs to.
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels =group_names, include_lowest=True)
df[['horsepower','horsepower-binned']].head()
```

```python
df["horsepower-binned"].value_counts()
```

```python
#%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
